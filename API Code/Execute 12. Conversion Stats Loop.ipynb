```python
# File to Load 
team_data_to_load = "Team_List.csv"
players_2018_data_to_load = "Offense_Player_List_2018.csv"

# Read Team and Player Data File and store into Pandas Data Frames
team_df = pd.read_csv(team_data_to_load)
players_2018_data = pd.read_csv(players_2018_data_to_load)
```

```python
#Conversions
#params to initiate loop through teams
year_list = ["2012", "2013", "2014", "2015", "2016", "2017", "2018"]
offensive_pos = ["TE", "WR", "RB", "QB", "FB", "K"]
master_player_list = players_2018_data["Player ID"].tolist()
team_list = team_df["Team ID"].tolist()
teamname_list = team_df["Name"].tolist()


pid_df = []
season_df = []
conv_def_attempts_df = []
conv_def_succ_df  =[]
conv_pass_attempts_df  =[]
conv_pass_succ_df  =[]
conv_rec_attempts_df  =[]
conv_rec_succ_df  =[] 
conv_rush_attempts_df  =[]
conv_rush_succ_df  =[]
conv_turnover_succ_df  = [] 



for a in range(len(year_list)):
    yr = year_list[a]
    for i in range (len(team_list)):
        print (str(i) + "||" + yr)
        url2 = "https://api.sportradar.us/nfl/official/trial/v5/en/seasons/" + yr + "/REG/teams/" + team_list[i] + "/statistics.json?api_key=" + key
        time.sleep(1)
        JSONContent2 = requests.get(url2).json()
        player_ct = len(JSONContent2["players"])
        for j in range (0, player_ct):
            pid = JSONContent2["players"][j]["id"]
            if pid in master_player_list:
                try: 
                    conv_def_attempts = JSONContent2["players"][0]["conversions"]["defense_attempts"] 
                    conv_def_succ =JSONContent2["players"][0]["conversions"]["defense_successes"]  
                    conv_pass_attempts =JSONContent2["players"][0]["conversions"]["pass_attempts"]  
                    conv_pass_succ =JSONContent2["players"][0]["conversions"]["pass_successes"]
                    conv_rec_attempts =JSONContent2["players"][0]["conversions"]["receive_attempts"]  
                    conv_rec_succ =JSONContent2["players"][0]["conversions"]["receive_successes"]  
                    conv_rush_attempts =JSONContent2["players"][0]["conversions"]["rush_attempts"]  
                    conv_rush_succ =JSONContent2["players"][0]["conversions"]["rush_successes"]
                    season_df.append(yr)
                    pid_df.append(pid)
                    conv_def_attempts_df.append(conv_def_attempts)
                    conv_def_succ_df.append(conv_def_succ)
                    conv_pass_attempts_df.append(conv_pass_attempts)
                    conv_pass_succ_df.append(conv_pass_succ)
                    conv_rec_attempts_df.append(conv_rec_attempts)
                    conv_rec_succ_df.append(conv_rec_succ)
                    conv_rush_attempts_df.append(conv_rush_attempts)
                    conv_rush_succ_df.append(conv_rush_succ)
                    conv_turnover_succ_df.append(0)
                except: 
                    try: 
                        conv_def_attempts =JSONContent2["players"][0]["conversions"]["defense_attempts"] 
                        conv_def_succ =JSONContent2["players"][0]["conversions"]["defense_successes"]  
                        conv_pass_attempts =JSONContent2["players"][0]["conversions"]["pass_attempts"]  
                        conv_pass_succ =JSONContent2["players"][0]["conversions"]["pass_successes"]
                        conv_turnover_succ =JSONContent2["players"][0]["conversions"]["turnover_successes"]  
                        conv_rush_attempts =JSONContent2["players"][0]["conversions"]["rush_attempts"]  
                        conv_rush_succ =JSONContent2["players"][0]["conversions"]["rush_successes"]
                        season_df.append(yr)
                        pid_df.append(pid)
                        conv_def_attempts_df.append(conv_def_attempts)
                        conv_def_succ_df.append(conv_def_succ)
                        conv_pass_attempts_df.append(conv_pass_attempts)
                        conv_pass_succ_df.append(conv_pass_succ)
                        conv_rec_attempts_df.append(0)
                        conv_rec_succ_df.append(0)
                        conv_rush_attempts_df.append(conv_rush_attempts)
                        conv_rush_succ_df.append(conv_rush_succ)
                        conv_turnover_succ_df.append(conv_turnover_succ)
                    except: 
                        print(sys.exc_info())  
                        continue 


data5 = {"Player Season": season_df,
        "Player ID": pid_df,
        "Converse Defense Attempts":conv_def_attempts_df,
        "Converse Defense Success":conv_def_succ_df,
        "Converse Pass Attempts":conv_pass_attempts_df,
        "Converse Pass Success":conv_pass_succ_df,
        "Converse Rec Attempts":conv_rec_attempts_df,
        "Converse Rec Success":conv_rec_succ_df,
        "Converse Rush Attempts":conv_rush_attempts_df,
        "Converse Rush Success":conv_rush_succ_df,
        "Converse Turnover Success":conv_turnover_succ_df

      }

player_conv_df = pd.DataFrame(data5) 
player_conv_df.head()

```
